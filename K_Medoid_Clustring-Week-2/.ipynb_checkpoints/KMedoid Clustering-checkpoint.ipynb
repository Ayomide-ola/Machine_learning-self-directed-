# K-medoids Clustering

# K-Medoids Clustering on Iris Data Set

```python
# import datasets from sklearn
from sklearn import datasets

# Load the dataset into an object iris
iris = datasets.load_iris()

# Display the iris dataset records
print(iris)
```

```python
# Create objects as X and y
X = []

# Initialise X with Iris data columns
X = iris.data

# Initialise y with the target column
y = iris.target
```

```python
print(X,y)
```

## Visualizing data

```python
import matplotlib.pyplot as plt
```

```python
# Visualizing data
fig, axes = plt.subplots(1, 2, figsize = (14, 6), dpi = 100)

# Scatter plot for 'Sepal length' and 'Sepal width'
axes[0].scatter(X[:, 0], X[:, 1], c = y)
axes[0].set_xlabel('Sepal Length')
axes[0].set_ylabel('Sepal Width')

# Scatter plot for 'Petal length' and 'Petal width'
axes[1].scatter(X[:, 2], X[:, 3], c = y)
axes[1].set_xlabel('Petal Length')
axes[1].set_ylabel('Petal Width')
```

## Develop Clusters using KMedoids clusterings using Petal length and Petal width.

```python
# Install the scikit-learn-extra library before the execution of next cell
!pip install scikit-learn-extra
```

```python
# Run this cell twice or more
from sklearn_extra.cluster import KMedoids

# Store 3rd and 4th column from X
D = X[:, 2:]
```

```python
print(D)
```

```python
import warnings
'warnings'(ignore)
```

```python
# Run this cell twice or more
from sklearn_extra.cluster import KMedoids

# Store 3rd and 4th column from X
D = X[:, 2:]

# Initialise the number of clusters
cluster_num = 3 # You can use elbow method to get the uptimal cluster number 

# model = k_medoids(k=cluster_num)
model   = KMedoids(n_clusters = 3, random_state = 0)
centers = model.fit(D)

# Display the centers
print(centers)
```

```python
# Display the attributes Petal length and Petal width
print(D)
```

```python
new_labels = model.predict(D)
# Plot the identified clusters and compare with our result
fig, axes = plt.subplots(1, 2, figsize = (14, 6), dpi = 100)
axes[0].scatter(D[:, 0], D[:, 1], c = y, cmap = 'gist_rainbow', edgecolor = 'k')
axes[1].scatter(D[:, 0], D[:, 1], c = new_labels, cmap = 'jet', edgecolor = 'k')
axes[0].set_xlabel('Petal length')
axes[0].set_ylabel('Petal width')
axes[1].set_xlabel('Petal length')
axes[1].set_ylabel('Petal width')
axes[0].set_title('Actual')
axes[1].set_title('Predicted')
```
